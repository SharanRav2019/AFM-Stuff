```python
import pandas as pd
import operator
import pylab
import numpy as np
from numpy import median
from matplotlib import pyplot as plt
from scipy.stats import normaltest
import statistics
from scipy import stats
from scipy.stats import kurtosis
from scipy.stats import skew

times = list()
times1 = list()
modes = list()

sd = pd.read_csv('a10r10_02.csv')
sd.t = sd.t/50000
num = 0;

plt.plot(sd.t*50000,sd.defl)
plt.show()

for x in range(750,len(sd.defl)-750,750):
    y = x-749
    data1 = statistics.stdev(sd.defl[y:x])
    data2 = statistics.stdev(sd.defl[(y+750):(x+750)])
    diff1 = data2 - data1;
    if(diff1 > 5.0e-11):
        num = num+1;
        print("we find a change point! difference:"+ str(diff1)+"!!!")
    modes.append(diff1);
    times.append(((y/5e4)+(x/5e4))/2)
    times1.append(int((x+y/2)))
    
print(num)
max_index1, max_value1 = max(enumerate(modes), key=operator.itemgetter(1))
print("Max mode"+str(modes[max_index1]))
print("Max time"+str(times[max_index1]))

plt.plot(times,modes,'o')
plt.show()
stdevs = np.asarray(modes)
time = np.asarray(times1)
tim = np.asarray(times)
msk = (stdevs>5.0e-11)
print (stdevs[msk])
print (tim[msk])
a = time[msk]
for i in a:
    plt.plot(sd.t[(i-50000):(i+50000)],sd.defl[(i-50000):(i+50000)])
    plt.show()
    print("Deflection change value:" + str(median(sd.defl[i:(i+5000)])-median(sd.defl[(i-5000):i])))
    
```
